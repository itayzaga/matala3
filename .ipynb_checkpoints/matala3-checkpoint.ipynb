```python
def anoname (file):
    newfile = ""
    lst = list()
    for line in file:
        line = line.strip()
        if line.count(':') >= 2:
            indexIDa = line.find('-')
            indexIDb = line[indexIDa:].find(':')
            if not line[(indexIDa + 2):(indexIDa + indexIDb)] in lst:
                lst.append(line[(indexIDa + 2):(indexIDa + indexIDb)])
            lineswitch = line
            newfile = newfile + lineswitch.replace(lineswitch[(indexIDa + 2):(indexIDa + indexIDb)] , str(lst.index(lineswitch[(indexIDa + 2):(indexIDa + indexIDb)]) + 1)) + "\n"
    ##print(newfile)
    #I will need the lst in the future so I will return it too.
    #In practice I also returned the newfile because it is the purpose of the function
    return [newfile,lst]

def orderfile (file):
    listMessages = list()
    
    respond = anoname(file)
    lst = respond[1]
    newfile = respond[0]
    #the will need the lst later
    
    for line in file:
        line = line.strip()
        if line.count(':') >= 2:
            dctForLine = dict()
            indexTime = line.find('-')
            dctForLine["datetime"] = line[:(indexTime - 1)]
            indexID = line[indexTime:].find(':')
            dctForLine["id"] = lst.index(line[(indexTime + 2):(indexTime + indexID)]) + 1
            dctForLine["text"] = line[(indexID + indexTime + 2):]
            listMessages.append(dctForLine)
    ##print(listMessages)
    return listMessages

def metaDa(file):
    dctMetadata = dict()
    relevanteLine = file[1]
    ##print(relevanteLine)
    indexNameA = relevanteLine.find('"')
    indexNameB = relevanteLine[(indexNameA + 1):].find('"')
    dctMetadata["chat_name"] = relevanteLine[(indexNameA + 1):(indexNameA + indexNameB + 1)]
    indexDate = relevanteLine.find('-')
    dctMetadata["creation_date"] = relevanteLine[:(indexDate - 1)]
    
    respond = anoname(file)
    lst = respond[1]
    newfile = respond[0]
    #the will need the lst later
    
    dctMetadata["num_of_participants"] = len(lst)
    indexCreator = relevanteLine.find("נוצרה על ידי")
    dctMetadata["creator"] = relevanteLine[(indexCreator + len("נוצרה על ידי") + 3):]
    ##print(dctMetadata)
    return dctMetadata
    
file = open('C:/Users/Moti/Desktop/python/text13.txt' , 'r' , encoding = 'utf-8')
file = file.readlines()
#I prefered to work with a list, not with a file for use many times 

dctTotal = {"messages" : orderfile(file) , "metadata" : metaDa(file)}
##print(dctTotal)

import json
dctJson = json.dumps(dctTotal)
indexNameA = file[1].find('"')
indexNameB = file[1][(indexNameA + 1):].find('"')
namegroup = file[1][(indexNameA + 1):(indexNameA + indexNameB + 1)]
fileJson = open('C:/Users/Moti/Desktop/python/%s.txt' % (namegroup) , 'a' , encoding = 'utf-8')
fileJson.write(dctJson)
fileJson.close()
```
